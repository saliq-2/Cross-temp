```python
import torch 
import torch.nn as nn

import torch.utils.data as data_utils
from torchvision import datasets, transforms
```

```python
class fashion(nn.Module):
    def __init__(self):
        super(fashion, self).__init__()

        self.layer1=nn.Linear(784,128)
        self.layer2=nn.Linear(128,64)
        self.layer3=nn.Linear(64,10)
        self.relu=nn.ReLU()
        self.softmax=nn.Softmax()
    def forward(self,input):
        input=input.view(input.shape[0],-1)
        input=self.relu(self.layer1(input))  
        input=self.relu(self.layer2(input))
        input=(self.layer3(input))


        #print(input[0])
        return input

        
```

```python
transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,), (0.5,))])
dataset_obj=datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)


print(dataset_obj)
```

```python
import torch
whole_dataset=len(dataset_obj.data)
split_ratio=0.8
train_set=int(whole_dataset*split_ratio)
print(train_set)
test_set=whole_dataset-train_set
print(test_set)
train_dataset, test_dataset = data_utils.random_split(dataset_obj, [train_set, test_set])


print("hello")
```

```python
train_loader=data_utils.DataLoader(dataset=train_dataset, batch_size=16, shuffle=True)
test_dataset_loader=data_utils.DataLoader(dataset=test_dataset, batch_size=16, shuffle=True)
```

```python
model=fashion()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
criterion=nn.CrossEntropyLoss()
images,labels=train_loader[0]
outputs=model(images)
print(outputs)

```

```python
def accuracy(outputs,labels):
    _,preds=torch.max(outputs,dim=1)
    return torch.tensor(torch.sum(preds==labels).item()/len(preds))
    
```

```python
import torch.optim as optim
epochs=10
for epichs in range(epochs):
    running_loss=0.0
    for images,labels in train_loader:
        optimizer.zero_grad()
        output=model(images)
        
        
        loss=criterion(output,labels)
        accuracy(output,labels)
        loss.backward()
        optimizer.step()
        running_loss+=loss.item()
    print(f"Epoch {epochs+1},Loss:{running_loss/len(train_loader)}")
    print(accuracy(output,labels).item())

        
        
```
